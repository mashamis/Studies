```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def median_filter(image, kernel_size=3):
    height, width = image.shape
    k = kernel_size // 2
    filtered_image = np.zeros_like(image)
    # Применение медианного фильтра
    for i in range(k, height - k):
        for j in range(k, width - k):
            neighborhood = image[i - k: i + k + 1, j - k: j + k + 1]
            filtered_image[i, j] = np.median(neighborhood)

    return filtered_image
```

```python
def minimum_filter(image, kernel_size=3):
    height, width = image.shape
    k = kernel_size // 2
    filtered_image = np.zeros_like(image)
    # Применение фильтра минимума
    for i in range(k, height - k):
        for j in range(k, width - k):
            # Выделение окрестности
            neighborhood = image[i - k: i + k + 1, j - k: j + k + 1]
            filtered_image[i, j] = np.min(neighborhood)
    return filtered_image
```

```python
def maximum_filter(image, kernel_size=3):
    height, width = image.shape
    k = kernel_size // 2
    filtered_image = np.zeros_like(image)
    # Применение фильтра максимума
    for i in range(k, height - k):
        for j in range(k, width - k):
            neighborhood = image[i - k: i + k + 1, j - k: j + k + 1]
            filtered_image[i, j] = np.max(neighborhood)

    return filtered_image
```

```python
image_min = cv2.imread('min_filter.png', cv2.IMREAD_GRAYSCALE)
image_max = cv2.imread('max_filter.png', cv2.IMREAD_GRAYSCALE)
image_median = cv2.imread('median.png', cv2.IMREAD_GRAYSCALE)
```

```python
def show(img, func, name:str):
    f, axarr = plt.subplots(nrows=1, ncols=4, figsize=(10, 10))
    axarr[0].imshow(img, cmap='gray')
    axarr[0].set_title('source image')
    axarr[1].imshow(func(img, 3), cmap='gray')
    axarr[1].set_title(f'{name} 3*3')
    axarr[2].imshow(func(img, 5), cmap='gray')
    axarr[2].set_title(f'{name} 5*5')
    axarr[3].imshow(func(img, 7), cmap='gray')
    axarr[3].set_title(f'{name} 7*7')
    plt.tight_layout()
    plt.show()
```

```python
show(image_median, median_filter, 'median')
show(image_max, maximum_filter, 'max')
show(image_min, minimum_filter, 'min')
```
